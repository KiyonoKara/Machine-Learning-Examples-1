Example using advertising data from "https://raw.githubusercontent.com/JWarmenhoven/ISLR-python/master/Notebooks/Data/Advertising.csv"

```python
import pandas as pd

# Load advertising data
url = "https://raw.githubusercontent.com/JWarmenhoven/ISLR-python/master/Notebooks/Data/Advertising.csv"
data = pd.read_csv(url)

print(data)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.lmplot(x="Newspaper", y="Sales", data=data, aspect=2)
plt.title("Scatter plot with linear regression line for newspaper and sales")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
X = data.drop(['Sales', 'Unnamed: 0'], axis=1)
y = data['Sales']

print(X, y)
```

```python
# TV and Sales
lr.fit(X[['TV']], y)

print("Estimated intercept: %.3f" %lr.intercept_)
print("Estimated coefficient: %.3f" %lr.coef_[-1])
```

```python
# All features and Sales
lr.fit(X, y)

# Coefficient table
coef_df = pd.DataFrame(X.columns)
coef_df.columns = ['Feature']
coef_df['Coefficient Estimate'] = pd.Series(lr.coef_)
print(coef_df)
```
