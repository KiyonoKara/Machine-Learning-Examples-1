```python
from sklearn.datasets import load_diabetes
data = load_diabetes()
X, y = data.data, data.target
print(data.DESCR)
```

```python
import pandas as pd

ld_dataframe = pd.DataFrame(data = X, columns = data.feature_names)
ld_dataframe['Target'] = y
```

Scatter plot between BMI and target, with linear regression

```python
import matplotlib.pyplot as plt
import seaborn as sns

# plt.scatter(X[0:,2], y)
plt.scatter(ld_dataframe['bmi'], y)
plt.xlabel("BMI")
plt.ylabel("Target")
plt.title("Scatter plot of BMI vs. Target")
plt.show()

sns.lmplot(data = ld_dataframe, x = 'bmi', y = 'Target', aspect = 2)
plt.title("Scatter plot with Linear Regression of BMI vs. Target")
plt.show()
```

Scatter plot between BP and Target, with linear regression

```python
# plt.scatter(X[0:,3], y)
plt.scatter(ld_dataframe['bp'], y)
plt.xlabel("BP")
plt.ylabel("Target")
plt.title("Scatter plot of BP vs. Target")
plt.show()

sns.lmplot(data = ld_dataframe, x = 'bp', y = 'Target', aspect = 2)
plt.title("Scatter plot with linear regression of BP vs. Target")
```
